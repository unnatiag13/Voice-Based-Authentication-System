```python
import librosa
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def extract_features(file):

    audio, sr = librosa.load(file, sr=16000)

    mfcc = librosa.feature.mfcc(
        y=audio,
        sr=sr,
        n_mfcc=13
    )

    mfcc = np.mean(mfcc.T, axis=0)

    return mfcc
```

```python
enroll_embedding = extract_features("enroll_voice.ogg")
test_embedding = extract_features("voice recording.ogg")

print("Enroll embedding:", enroll_embedding)
print("Test embedding:", test_embedding)
```

```python
score = cosine_similarity(
    enroll_embedding.reshape(1,-1),
    test_embedding.reshape(1,-1)
)

print("Similarity Score:", score[0][0])
```

```python
threshold = 0.85

if score > threshold:
    print("✅ Access Granted (Same Speaker)")
else:
    print("❌ Access Denied (Different Speaker)")
```
